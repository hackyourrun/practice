```python
import pandas as pd
```

```python
url = "http://steamtownmarathon.com/wp-content/uploads/2019/10/Steamtown-Final-Report-10-16-19-Update.xls"
timedelta_cols = ['18 MILE SPLIT', 'CHIP TIME', 'GUN TIME', 'GUN PACE']
df = pd.read_excel(url, converters={col:str for col in timedelta_cols})
```

```python
for col in timedelta_cols:
    df[col] = pd.to_timedelta(df[col])
```

```python
df['18_mile_pace'] = pd.Series(df['18 MILE SPLIT'] / 18)  # makes a new column with pace for 18 miles 
df['last_8.2_mile_pace'] = pd.Series((df['CHIP TIME'] - df['18 MILE SPLIT'])/8.2)  # makes a new column with pace for rest of race
df['change_in_pace'] = pd.Series(abs(df['18_mile_pace'] - df['last_8.2_mile_pace']))  # makes a new column for difference in pace
```

```python
df.sort_values('change_in_pace') #sorts based on change in pace from the split
```
